```python
# dataframe

data={
    "Name":["alice","bob","charlie","david","eve","Alice"],
    "age":[25,35,46,53,54,32],
    "department":["HR","IT","Finance","HR","IT","Finance"],
    "salary":[50000,60000,70000,55000,np.nan,75000]
}
```
